```python
violent_crimes = df.convert_objects(convert_numeric=True)
violent_crimes['Date'] = pd.to_datetime(violent_crimes['Date'])
```

### have datetime column with m/d/y and want to show the cumulative count across all the years for each m/d

```python
sexual_assault = violent_crimes[(violent_crimes['Primary Type'] == 'CRIM SEXUAL ASSAULT')]

# sexual_assault['Date'] = pd.DatetimeIndex(sexual_assault['Date'])

dates = pd.DatetimeIndex(sexual_assault['Date'])
counts = sexual_assault.groupby(dates.month).cumsum()

# errors = sexual_assault.groupby(dates.month).std()
# ax=sexual_assault.groupby(dates.day).cumsum().plot(yerr=errors, kind='bar')
# ax.set_xlabel("months")
# ax.set_ylabel("gross")
```
